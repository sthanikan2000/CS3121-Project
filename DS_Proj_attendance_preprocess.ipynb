```python
import pandas as pd
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('attendance.csv')
df.shape
```

```python
# df['date'].isin(['\\N', '0000-00-00']).sum() # output zero
mask=df['Hourly_Time'].isin(['\\N'])
df.loc[mask, 'Hourly_Time'] = (pd.to_timedelta(df.loc[mask, 'out_time']) - pd.to_timedelta(df.loc[mask, 'in_time'])).dt.total_seconds() / 3600 #mask=(df["date"] != df["out_date"]) # 10111 records out of 224057(4.5% of records)
# df.loc[mask, 'out_date'] = df.loc[mask, 'date']
df[mask]
```

```python
df.loc[df['out_time'].str.startswith("24"), "out_time"] = df.loc[df['out_time'].str.startswith("24"), "out_time"].str.replace("^24", "00", regex=True)
```

```python
df[mask]
```

```python
mask = df['out_date'].isin(['\\N', '0000-00-00'])  # Only one record
df[mask]
# we don't need pre processing on this
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.nunique()
```

```python
# Convert columns to datetime type
in_time = pd.to_timedelta(df['in_time'])
shift_start = pd.to_timedelta(df['Shift_Start'])
# Calculate late minutes
df['Late_minutes'] = (in_time-shift_start).dt.total_seconds() / 60
df['Late_minutes'] = df['Late_minutes'].clip(lower=0)
df.head(20)
```

```python
df.drop(["id",'project_code', 'date',"out_date", "out_time","in_time","Shift_Start","Shift_End"], axis=1, inplace=True)
```

```python
df['Hourly_Time'] = df['Hourly_Time'].astype(float)
```

```python
df.info()
```

```python
# Group by 'Employee_No' and calculate the average
averages = df.groupby('Employee_No').mean()
averages
```

```python
averages.describe()
```

```python
averages.to_csv('attendance_preprocessed_200635E.csv', index=True)
```
